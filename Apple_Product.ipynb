# **Top 10 Apple iphones that sales in India**
Lets find top 10 highest-rated iPhones in India in flipkart. It will help understand what kind of iPhones are liked the most in India.



```python
highest_rated=df.sort_values(by=["Star Rating"],ascending=False).head(10)
highest_rated=highest_rated.head(10)
highest_rated["Product Name"]
```

Now lets have a look at the number of ratings of the highest-rated iPhones from flipkart








```python
highest_rated['Product Name'].value_counts()
```

```python
iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]

plt.figure(figsize=(10, 6))
sns.barplot(x=label, y=counts,palette="Set2")
plt.title("Number of Ratings of Highest Rated iPhones in India")
plt.xlabel("iPhone Models")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Now lets have a look at the relationship between the sale price of iPhones and their ratings on Flipkart:

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Number Of Ratings", y="Sale Price", size="Discount Percentage", sizes=(10, 200))
sns.regplot(data=df, x="Number Of Ratings", y="Sale Price", scatter=False)
plt.title("Sale Price vs Number of Ratings with Discount Percentage")
plt.show()
```

There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more in India.

Now lets have a look at the relationship between the discount percentage on iPhones on Flipkart and the number of ratings:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="Number Of Ratings", y="Discount Percentage", scatter_kws={"s": df["Sale Price"] * 0.01})
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones")
plt.show()
```

**Summary:** There is a linear relationship between the discount percentage on iPhones on Flipkart link textand the number of ratings. It means iPhones with high discounts are sold more in India
